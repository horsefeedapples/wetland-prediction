```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
```

```python
p = sns.jointplot(data=df, x="sDEM", y="DSM", hue="Wetland")
p.fig.suptitle("sDEM vs DSM between Wetlands and Nonwetlands")
p.fig.tight_layout()
p.fig.subplots_adjust(top=0.95)
```
