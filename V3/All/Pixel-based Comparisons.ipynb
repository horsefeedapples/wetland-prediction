```python
model = ["GBM", "LGBM", "XGBM"]
accuracy = [0.8301, 0.8127, 0.8320]
specificity = [0.7469, 0.7445, 0.7666]
sensitivity = [0.8839, 0.8569, 0.8744]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(20, 10)) 

ax1.bar(model, accuracy, tick_label=model)
ax1.set_title("Accuracy for Gradient Boosting Models")
ax1.set_xlabel("GB Model")
ax1.set_ylabel("Accuracy")
ax1.set_ylim([.7, .89])

ax2.bar(model, sensitivity, tick_label=model)
ax2.set_title("Sensitivity for Gradient Boosting Models")
ax2.set_xlabel("GB Model")
ax2.set_ylabel("Sensitivity")
ax2.set_ylim([.7, .89])

ax3.bar(model, specificity, tick_label=model)
ax3.set_title("Specificity for Gradient Boosting Models")
ax3.set_xlabel("GB Model")
ax3.set_ylabel("Specificity")
ax3.set_ylim([.70, .85])
```

```python
sites = ["CastleBay", "MasonHT", "MasonLT", "Maysville", "RiverRoad", "StJames", "SurfCity", "Topsail", "All"]

x = np.arange(len(sites))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x-width, GBM_accuracy, width, label='GBM')
rects2 = ax.bar(x, LGBM_accuracy, width, label='LGBM')
rects3 = ax.bar(x+width, XGBM_accuracy, width, label='XGBM')

# Add some text for labels, title and custom x-axis tick labels, etc.
font_dict = {'fontsize': 8, 'fontweight': 'normal', 'verticalalignment':'top'}
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy for different wetland sites')
ax.set_xticks(x + width / 3)
ax.set_xticklabels(sites, fontdict = font_dict)
ax.set_xlabel('Sites', labelpad=15)
ax.legend()

fig.tight_layout()
ax.set_ylim([.5, 1])

plt.show()
```
