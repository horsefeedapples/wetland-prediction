```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import os
import pandas as pd
import seaborn as sns

%matplotlib inline
sns.set_style('whitegrid')
sns.set(rc={"figure.dpi":70, 'savefig.dpi':70}) #defining dpi setting
sns.set_context('notebook')
sns.set_style("ticks")
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
```

```python
dfes = pd.read_csv('../RR_Quan_2m_dataset.csv')
dfes['FID_RiverRd_AOI'] = np.where(dfes['NWI_bi']=='wetland', 1, 0)
```

```python
def vplt(columnName):
    fig, ax = plt.subplots(2, 2, figsize=(12,9))
    sns.violinplot(x="NWI_bi", y=f"{columnName}_MIN", data=dfes, ax=ax[0,0])
    sns.violinplot(x="NWI_bi", y=f"{columnName}_MAX", data=dfes, ax=ax[0,1])
    sns.violinplot(x="NWI_bi", y=f"{columnName}_MEAN", data=dfes, ax=ax[1,0])
    sns.violinplot(x="NWI_bi", y=f"{columnName}_STD", data=dfes, ax=ax[1,1])
    fig.suptitle(f"{columnName} Distribution Plot")
```

```python
for(columnName, columnData) in dfes.iteritems():
    if("MIN" in columnName):
        columnName = columnName.split('_')[0]
        vplt(columnName)
```
