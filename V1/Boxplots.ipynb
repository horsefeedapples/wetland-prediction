```python
def stacked_boxplots(dfes, hue, statistic, title):
    fig, ax = plt.subplots(3, 2, figsize=(12,16))
    sns.set_style("white")


    sns.boxplot(data=dfes, ax=ax[0,0], x="wetland", y=f"{statistic}_MEAN", hue=hue).set_title(f"{title} Boxplot: Mean vs Wetland Type")

    sns.boxplot(data=dfes,ax=ax[0,1], x="wetland", y=f"{statistic}_STD", hue=hue).set_title(f"{title} Boxplot: Standard Deviation vs Wetland Type")

    sns.boxplot(data=dfes,ax=ax[1,0], x="wetland", y=f"{statistic}_RANGE", hue=hue).set_title(f"{title} Boxplot: Range vs Wetland Type")

    sns.boxplot(data=dfes, ax=ax[1,1], x="wetland", y=f"{statistic}_MIN", hue=hue).set_title(f"{title} Boxplot: Minimum vs Wetland Type")

    sns.boxplot(data=dfes, ax=ax[2,0], x="wetland", y=f"{statistic}_MAX", hue=hue).set_title(f"{title} Boxplot: Maximum vs Wetland Type")

    sns.boxplot(data=dfes, ax=ax[2,1], x="wetland", y=f"{statistic}_SUM", hue=hue).set_title(f"{title} Boxplot: Sum vs Wetland Type")
```

```python
dem = stacked_boxplots(dfes, "wetland", "qu_dem", "DEM")
```
