```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import os
import pandas as pd
import seaborn as sns
import SeabornFig2Grid as sfg

%matplotlib inline
sns.set_style('whitegrid')
sns.set(rc={"figure.dpi":70, 'savefig.dpi':70}) #defining dpi setting
sns.set_context('notebook')
sns.set_style("ticks")
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
```

```python
dfes = pd.read_csv('RiverRoad_Quan_RF.csv')
```

```python
def jpt(dfes, x, y, hue, title):
    g = sns.jointplot(data=dfes, x=x, y=y, hue=hue)
    g.fig.suptitle(title)
```

```python
g1 = jpt(dfes, "qu_dem_MIN", "qu_dem_MAX", "wetland", "Wetland Scatterplot by DEM")
g2 = jpt(dfes, "qu_dem_MEAN", "qu_dem_STD", "wetland", "Wetland Scatterplot by DEM")
```

```python
g1 = jpt(dfes, "qu_curpl_MIN", "qu_curpl_MAX", "wetland", "Wetland Scatterplot by Curvature Standard")
g2 = jpt(dfes, "qu_curpl_MEAN", "qu_curpl_STD", "wetland", "Wetland Scatterplot by Curvature Standard")
```

```python
g1 = jpt(dfes, "qu_floAcu_MIN", "qu_floAcu_MAX", "wetland", "Wetland Scatterplot by Flow Accumulation")
g2 = jpt(dfes, "qu_floAcu_MEAN", "qu_floAcu_STD", "wetland", "Wetland Scatterplot by Flow Accumulation")
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(16,6))
ax1.set_title("Flow Accumulation for Wetlands")
sns.scatterplot(data=dfes[dfes['wetland']=='wetland'], x="qu_floAcu_MEAN", y="qu_floAcu_STD", hue="wetland", palette=["darkorange"], ax=ax1)
ax2.set_title("Flow Accumulation for Non-Wetlands")
sns.scatterplot(data=dfes[dfes['wetland']=='nonwetland'], x="qu_floAcu_MEAN", y="qu_floAcu_STD", hue="wetland", palette=["blue"], ax=ax2)
```
