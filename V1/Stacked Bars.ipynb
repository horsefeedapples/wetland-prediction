```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import seaborn as sns

%matplotlib inline
plt.rcParams['figure.dpi'] = 70 #display 70 dpi in Jupyter Notebook, may consider100 dpi 
plt.rcParams['savefig.dpi'] = 300 #define 300 dpi for saving figures
sns.set_style('whitegrid')
sns.set(rc={"figure.dpi":70, 'savefig.dpi':300}) #defining dpi setting
sns.set_context('notebook')
sns.set_style("ticks")
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
```

```python
dfes = pd.read_csv('RiverRoad_Quan_RF.csv')
#only wetland
dfes = dfes[dfes['wetland']=='wetland']
```

```python
def stacked_histograms(dfes, statistic, hue, title):
    fig, ax = plt.subplots(2, 3, figsize=(20, 12))
    sns.set_style("white")

    sns.histplot(data=dfes, ax=ax[0,0], stat="count", multiple="stack", x=f"{statistic}_MEAN", hue=hue, element="bars", legend=True).set_title(f"{title} Mean vs Wetland Type")

    sns.histplot(data=dfes,ax=ax[0,1], stat="count", multiple="stack", x=f"{statistic}_STD", hue=hue, element="bars", legend=True).set_title(f"{title} Standard Deviation vs Wetland Type")

    sns.histplot(data=dfes,ax=ax[0,2], stat="count", multiple="stack", x=f"{statistic}_RANGE", hue=hue, element="bars", legend=True).set_title(f"{title} Range vs Wetland Type")

    sns.histplot(data=dfes, ax=ax[1,0], stat="count", multiple="stack", x=f"{statistic}_MIN", hue=hue, element="bars", legend=True).set_title(f"{title} Minimum vs Wetland Type")

    sns.histplot(data=dfes, ax=ax[1,1], stat="count", multiple="stack", x=f"{statistic}_MAX", hue=hue, element="bars", legend=True).set_title(f"{title} Maximum vs Wetland Type")

    sns.histplot(data=dfes, ax=ax[1,2], stat="count", multiple="stack", x=f"{statistic}_SUM", hue=hue, element="bars", legend=True).set_title(f"{title} Sum vs Wetland Type")
```

```python
dem = stacked_histograms(dfes, "qu_dem", "wetland_type", "DEM")
```

```python
smdem = stacked_histograms(dfes, "qu_smdem", "wetland_type", "Smooth High-Res DEM")
```

```python
hyddem = stacked_histograms(dfes, "qu_hyddem", "wetland_type", "High-Conditioned High-Res DEM")
```

```python
dsm = stacked_histograms(dfes, "qu_dsm", "wetland_type", "DSM")
```
